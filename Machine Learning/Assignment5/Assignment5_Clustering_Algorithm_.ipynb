```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Load and preprocess data droping the species columns
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

**A) KMeans Clustering**

KMeans Clustering is an unsupervised algorithm that partitions the dataset into K distinct non-overlapping clusters.

It works as follows:

* Randomly initialize K centroids.

* Assign each point to the nearest centroid.

* Recalculate centroids based on current cluster memberships.

* Repeat until convergence (no change in assignments or centroids).

**Why it's suitable for the Iris dataset:**

The Iris dataset contains numerical features (like sepal/petal length and width), which makes it a good candidate for distance-based algorithms like KMeans.


```python
#applying Kmeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)

#visualizing
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=kmeans_labels, palette='viridis')
plt.title("KMeans Clustering on Iris Dataset ")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend(title='Cluster')
plt.show()
```

**B) Hierarchical Clustering**

builds a hierarchy of clusters using either a bottom-up (agglomerative) or top-down (divisive) approach.

* In agglomerative clustering, each data point starts as its own cluster.

* Clusters are merged based on distance (e.g., Euclidean), forming a dendrogram.

* The dendrogram is cut at a certain height to form final clusters.

**Why it's suitable for the Iris dataset:**

Hierarchical clustering doesn't require pre-specifying the number of clusters and offers an interpretable tree-like structure (dendrogram), which is useful for understanding the natural grouping in the data.

```python
# B) Hierarchical Clustering
Z = linkage(X_scaled, 'ward')

# Visualizing Hierarchical Clustering (Dendrogram)
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
